### If the data distributed other than linear we use ploynomial linear regression

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#taking data
np.random.seed(42)
X = np.linspace(-5, 5, 100).reshape(-1, 1)        # Independent variable
y = X**2 + np.random.randn(100, 1) * 3    
```

```python
#seeing distribution of the data
plt.scatter(X, y, color='blue', label='Nonlinear data (y = x² + noise)')
plt.title("Generated Nonlinear Dataset")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

- if we use any linear regression model we get more error to reduce such error we use polynomial regression model

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=3,include_bias=True) #we get three variable from a single variable X that is X^0 ,X^1 ,X^2
```

```python
x_train_poly = poly.fit_transform(x_train)
x_test = poly.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()

model.fit(x_train_poly,y_train)
```

```python
from sklearn.metrics import r2_score

score =  r2_score(y_test,model.predict(x_test))
score
```

```python
plt.scatter(x_train,model.predict(x_train_poly))
plt.scatter(x_train,y_train)
```
